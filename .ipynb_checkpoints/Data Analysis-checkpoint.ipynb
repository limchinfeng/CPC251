```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('climate.csv')  # Replace 'your_dataset.csv' with your actual dataset filename

# Explore the dataset
print(df.head())  # Display the first few rows of the dataset
print(df.info())  # Get summary information about the dataset

# Perform data analysis

# Descriptive statistics
summary_stats = df.describe()  # Calculate summary statistics
print(summary_stats)
```

```python
# Bar plot of outcome counts before resampling
outcome_counts = df['outcome'].value_counts()
plt.bar(outcome_counts.index, outcome_counts.values)
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.title('Bar Plot of Outcome Counts (Before Resampling)')
plt.xticks([0, 1])  # Set x-axis ticks for binary classes
plt.show()
```
